```python
# MLP for Pima Indians Dataset with 10-fold cross validation via sklearn
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
import os
import numpy

os.chdir('C:/deep_learning') # This is where the input dataset is stored
os.getcwd()
```

```python
# Function to create model, required for KerasClassifier
def create_model():
    # create model
    in_model = Sequential()
    in_model.add(Dense(12, input_dim=8, activation='relu'))
    in_model.add(Dense(8, activation='relu'))
    in_model.add(Dense(1, activation='sigmoid'))
    # Compile model
    in_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return in_model

# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# create model
# The KerasClassifier and KerasRegressor classes in Keras take an argument 
# build_fn which is the name of the function to call to get your model.
model = KerasClassifier(build_fn=create_model, epochs=50, batch_size=10, verbose=1)
```

```python
# evaluate using 5-fold cross validation
kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=seed)
```

```python
# cross_val_score: a scikit-learn function to evaluate our model using the cross-validation scheme and print the results
results = cross_val_score(model, X, Y, cv=kfold)
print(results)
```
