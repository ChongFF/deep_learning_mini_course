```python
# MLP for Pima Indians Dataset with 10-fold cross validation via sklearn
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
import os
import numpy
import datetime as dt

os.chdir('C:/deep_learning') # This is where the input dataset is stored
os.getcwd()
```

```python
# Function to create model, required for KerasClassifier
def create_model(optimizer='rmsprop', init='glorot_uniform'):
    # create model
    in_model = Sequential()
    in_model.add(Dense(12, input_dim=8, activation='relu'))
    in_model.add(Dense(8, activation='relu'))
    in_model.add(Dense(1, activation='sigmoid'))
    # Compile model
    in_model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return in_model

# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# create model
# The KerasClassifier and KerasRegressor classes in Keras take an argument 
# build_fn which is the name of the function to call to get your model.
model = KerasClassifier(build_fn=create_model, verbose=1)
```

```python
# grid search epochs, batch size and optimizer
optimizers = ['rmsprop', 'adam']
init = ['glorot_uniform', 'normal', 'uniform']
epochs = [10, 15, 20]
batches = [5, 10, 20]
param_grid = dict(optimizer=optimizers, epochs=epochs, batch_size=batches, init=init)
grid = GridSearchCV(estimator=model, param_grid=param_grid)
```

```python
grid_result = grid.fit(X, Y)
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
